## Results
Training prints 3 audio scenarios at the end:
- **high_energy**: should show dense wall glyphs, bright
- **low_energy**: should show sparse/dim corridor
- **onset_burst**: should show flash characters

If the frames look different across scenarios, audio conditioning is working.
